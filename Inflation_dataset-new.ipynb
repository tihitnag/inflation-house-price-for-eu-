```python
import pandas as pd
import numpy as np


```

```python

file_path = 'Inflation-data______new (1).xlsx'
df = pd.read_excel(file_path, sheet_name=0)

# df.columns.tolist()
```

# Dropping columns for years earlier than 2011 and later than 2021

```python
df = df.drop(columns=[col for col in df.columns if isinstance(col, int) and (col < 2011 or col > 2021)])

# Displaying column names after keeping only years from 2010 to 2021
print("\n=== Columns after keeping only years 2010 - 2021 ===")
print(list(df.columns))
```

```python
countries_to_keep = [
    "Belgium", "Bulgaria", "Czechia", "Denmark", "Germany", "Estonia", "Ireland",
    "Greece", "Spain", "France", "Italy", "Latvia",
    "Lithuania", "Luxembourg", "Hungary", "Malta", "Netherlands", "Austria",
    "Poland", "Portugal", "Slovenia", "Slovakia", "Finland",
    "Sweden", "Iceland", "Norway"
]

# Filtrowanie tylko wybranych krajów
df = df[df["Country"].isin(countries_to_keep)]
```

```python
print("\n=== Lista unikalnych krajów ===")
print(df["Country"].unique())
```

```python
df
```

# Resetting the DataFrame index to start from 0

```python
df = df.reset_index(drop=True)
```

```python
df
```

# Check for missing values

```python
print("\n=== Number of missing values in each column ===")
print(df.isna().sum())
```

```python
print("\n=== Data types ===")
print(df.dtypes)
```

```python
df = df.drop(columns=["Country Code", "IMF Country Code", "Indicator Type"])
```

```python
df
```

```python
# Calculate average, min and max inflation for each country (2010–2021)
inflation_stats = df.groupby("Country")[list(range(2010, 2021 + 1))].agg(['mean', 'min', 'max'])

# Flatten multi-index columns for readability
inflation_stats.columns = ['_'.join(col).strip() for col in inflation_stats.columns]

print("\n=== Average, Minimum and Maximum Inflation per Country (2010-2021) ===")
print(inflation_stats)
```
